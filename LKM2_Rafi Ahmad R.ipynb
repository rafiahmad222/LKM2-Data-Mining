```python
data =pd.read_csv('iris_dataset.csv')
```

```python
setosa = data[data['Species_name'] == 'Setosa'].values[:, 1:]
print("Setosa:")
print(setosa[: ])

versicolor = data[data['Species_name'] == 'Versicolor'].values[:, 1:]
print("\nVersicolor:")
print(versicolor[: ])

virginica = data[data['Species_name'] == 'Verginica'].values[:, 1:]
print("\nVirginica:")
print(virginica[: ])

unknown = data[data['Species_name'] == 'unknown'].values[:, 1:]
print("\nUnknown:")
print(unknown[: ])

```

```python
# Memisahkan data berdasarkan kelas
setosa = data[data['Species_name'] == 'Setosa']
versicolor = data[data['Species_name'] == 'Versicolor']
verginica = data[data['Species_name'] == 'Verginica']
unknown = data[data['Species_name'] == 'unknown']

plt.figure(figsize=(15, 12))

# Sepal length dan petal width
plt.subplot(2, 2, 4)
plt.scatter(setosa['Sepal_length'], setosa['Petal_width'], color='red', label='Setosa')
plt.scatter(versicolor['Sepal_length'], versicolor['Petal_width'], color='green', label='Versicolor')
plt.scatter(verginica['Sepal_length'], verginica['Petal_width'], color='blue', label='Virginica')
plt.scatter(unknown['Sepal_length'], unknown['Petal_width'], color='yellow', label='Unknown')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.title('Sepal Length vs Petal Width')
plt.legend()

# Sepal width dan petal length
plt.subplot(2, 2, 3)
plt.scatter(setosa['Sepal_width'], setosa['Petal_length'], color='red', label='Setosa')
plt.scatter(versicolor['Sepal_width'], versicolor['Petal_length'], color='green', label='Versicolor')
plt.scatter(verginica['Sepal_width'], verginica['Petal_length'], color='blue', label='Virginica')
plt.scatter(unknown['Sepal_width'], unknown['Petal_length'], color='yellow', label='Unknown')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Length')
plt.title('Sepal Width vs Petal Length')
plt.legend()

# Sepal length dan petal length
plt.subplot(2, 2, 1)
plt.scatter(setosa['Sepal_length'], setosa['Petal_length'], color='red', label='Setosa')
plt.scatter(versicolor['Sepal_length'], versicolor['Petal_length'], color='green', label='Versicolor')
plt.scatter(verginica['Sepal_length'], verginica['Petal_length'], color='blue', label='Virginica')
plt.scatter(unknown['Sepal_length'], unknown['Petal_length'], color='yellow', label='Unknown')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Sepal Length vs Petal Length')
plt.legend()

# Sepal width dan petal width
plt.subplot(2, 2, 2)
plt.scatter(setosa['Sepal_width'], setosa['Petal_width'], color='red', label='Setosa')
plt.scatter(versicolor['Sepal_width'], versicolor['Petal_width'], color='green', label='Versicolor')
plt.scatter(verginica['Sepal_width'], verginica['Petal_width'], color='blue', label='Virginica')
plt.scatter(unknown['Sepal_width'], unknown['Petal_width'], color='yellow', label='Unknown')
plt.xlabel('Sepal width')
plt.ylabel('Petal Width')
plt.title('Sepal width vs Petal Width')
plt.legend()

# Untuk menghasilkan output
plt.tight_layout()
plt.show ()
```

```python
# Memilih data berkelas Setosa, Versicolor, dan Virginica
setosa_data = data[data['Species_name'] == 'Setosa']
versicolor_data = data[data['Species_name'] == 'Versicolor']
verginica_data = data[data['Species_name'] == 'Verginica']

for _, row in setosa_data.iterrows():
    plt.plot(range(1, 5), row[['Petal_width', 'Petal_length', 'Sepal_width', 'Sepal_length']], color='red', alpha=0.5)

for _, row in versicolor_data.iterrows():
    plt.plot(range(1, 5), row[['Petal_width', 'Petal_length', 'Sepal_width', 'Sepal_length']], color='green', alpha=0.5)

for _, row in verginica_data.iterrows():
    plt.plot(range(1, 5), row[['Petal_width', 'Petal_length', 'Sepal_width', 'Sepal_length']], color='blue', alpha=0.5)

plt.xticks(range(1, 5), ['Petal Width', 'Petal Length', 'Sepal Width', 'Sepal Length'])
plt.xlabel('Features')
plt.ylabel('Length (cm)')
plt.title('Paralel Coordinat Plot')

plt.legend(['Setosa', 'Versicolor', 'Virginica'])

plt.grid(True)
plt.show()
```
